```python
import pandas as pd
```

```python
file = "Resources/purchase_data.csv"

purchases_data = pd.read_csv(file)
purchases_data.head(20)
```

```python
#total = purchases_data.shape[0]
total = purchases_data["SN"].nunique()

data = [[total]]
total_players = pd.DataFrame(data, columns = ["Total Players"])
# total_players = pd.DataFrame({"Total Players": total})
total_players

 
# Create the pandas DataFrame
```

```python
unique_items = purchases_data["Item ID"].nunique()

average_price = purchases_data["Price"].mean()
final_price = round(average_price, 2)


total_purchases = purchases_data["Price"].count()

total_revenue = purchases_data["Price"].sum()


together = unique_items, final_price, total_purchases, total_revenue

purchases_analysis = pd.DataFrame(data = [together], columns = ["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"], index=[0])
purchases_analysis["Average Price"] = purchases_analysis["Average Price"].map("${:,.2f}".format)
purchases_analysis["Total Revenue"] = purchases_analysis["Total Revenue"].map("${:,.2f}".format)
purchases_analysis
```

```python
total_genders = pd.DataFrame(purchases_data.drop_duplicates(["SN"]))
grouped_genders = total_genders.groupby(["Gender"])
count_genders = pd.DataFrame({"Total Count": grouped_genders["Gender"].count(), "Percentage of Players": grouped_genders["Gender"].count()/total * 100})
count_genders["Percentage of Players"] = count_genders["Percentage of Players"].map("{:.2f}%".format)
count_genders.head()
```

```python
purchases = pd.DataFrame(purchases_data)
gender_purchases = purchases.groupby(["Gender"])

gender_purchases_summary = pd.DataFrame ({"Purchase Count": gender_purchases["Price"].count(), "Average Purchase Price": gender_purchases["Price"].mean(),"Total Purchase Value": gender_purchases["Price"].sum(), "Avg Total Purchases per person": gender_purchases["Price"].sum()/ count_genders["Total Count"]})
gender_purchases_summary["Average Purchase Price"] = gender_purchases_summary["Average Purchase Price"].map("${:,.2f}".format)
gender_purchases_summary["Total Purchase Value"] = gender_purchases_summary["Total Purchase Value"].map("${:,.2f}".format) 
gender_purchases_summary["Avg Total Purchases per person"] = gender_purchases_summary["Avg Total Purchases per person"].map("${:,.2f}".format)
gender_purchases_summary.head()
```

```python
bins = [0, 5, 10, 15, 20, 25, 30, 35, 40]
labled = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchases_data["Age Groups"] = pd.cut(purchases_data["Age"], bins, labels=labled)
age_grouped = purchases_data.groupby(["Age Groups"])
age_count = age_grouped["SN"].nunique()
age_percent = age_count / total_players * 100
age_data = pd.DataFrame({"Total Count": age_count, "Percentage of Players": age_percent})

```

```python
age_perchase = age_grouped["Purchase ID"].count()
avg_age = age_grouped["Price"].mean()
total_age_purchases = age_grouped["Price"].sum()
avg_total_purchase = total_age_purchases / age_count * 100
age_analysis = pd.DataFrame ({"Purchase Count": age_perchase, "Average Purchase Price": avg_age, "Total Purchase Value": total_age_purchases, "Avg Total Purchases per person": avg_total_purchase})
age_analysis["Average Purchase Price"] = age_analysis["Average Purchase Price"].map("${:,.2f}".format)
age_analysis["Total Purchase Value"] = age_analysis["Total Purchase Value"].map("${:,.2f}".format)
age_analysis["Avg Total Purchases per person"] = age_analysis["Avg Total Purchases per person"].map("${:,.2f}".format)
age_analysis
```

```python
spenders = pd.DataFrame(purchases_data)
top_grouped = spenders.groupby(["SN"])
top = pd.DataFrame({"Purchases Count": top_grouped["Purchase ID"].count(), "Average Purchase Price": top_grouped["Price"].mean(), "Total Purchase Value": top_grouped["Price"].sum()})
top = top.sort_values(["Total Purchase Value"], ascending= False)
top["Average Purchase Price"] = top["Average Purchase Price"].map("${:,.2f}".format)
top["Total Purchase Value"] = top["Total Purchase Value"].map("${:,.2f}".format)
top.head()
```
