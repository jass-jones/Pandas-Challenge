```python
import pandas as pd
```

```python
file = "Resources/schools_complete.csv"
file_two = "Resources/students_complete.csv"
school_data = pd.read_csv(file)
student_data = pd.read_csv(file_two)
merge = pd.merge(school_data, student_data, how= "left", on= ["school_name", "school_name"])
merge.head()
```

```python
total_schools = len(merge["school_name"].unique())
total_students = merge["student_name"].nunique()
budget = school_data["budget"].sum()
avg_math = merge["math_score"].mean()
avg_reading = merge["reading_score"].mean()
passing_math = len(merge[merge["math_score"] >= 70])
math_percent = passing_math /total_students * 100
passing_reading = len(merge[merge["reading_score"] >= 70])
reading_percent = passing_reading / total_students * 100
together = total_schools, total_students, budget, avg_math, avg_reading, math_percent, reading_percent

district_summary = pd.DataFrame(data = [together], columns=["Toal Schools", "Total Students", "Total Budget", "Average Math Score", "Avergae Reading Score", "%Passing Math", "%Passing Reading"])
district_summary
```

```python
merge_data = pd.DataFrame(merge)
school_name = merge_data.groupby(["school_name"])
school_type = school_name["type"].first()
number_of_students = school_name["Student ID"].count()
school_budget = merge_data.groupby(["budget"]).first()
student_budget = school_budget / number_of_students
avgMath = merge_data["math_score"].mean()
avgRead = merge_data["reading_score"].mean()
passMath = merge[merge["math_score"] >= 70].groupby(["school_name"])["math_score"].count()
percentMath = passMath / number_of_students * 100
passRead = merge[merge["reading_score"] >= 70].groupby(["school_name"])["reading_score"].count()
percentRead = passRead / number_of_students * 100
overall = (percentMath + percentRead) / 2

school_summary = pd.DataFrame({"School Type": school_type, "Total Students": number_of_students, 
"Total School Budget":school_budget, "Per Student Budget": student_budget, "Average Math Score": avgMath, 
"Average Reading Score": avgRead, "% Passing Math": percentMath, "% Passing Reading": percentRead, 
"% Overall Passing Math and Raeding": overall})
school_summary.head()


```

```python
grade9 = merge.loc[merge["grade"] == "9th"].groupby(["school_name"]) ["math_score"].mean()
grade10 = merge.loc[merge["grade"] == "10th"].groupby(["school_name"]) ["math_score"].mean()
grade11 = merge.loc[merge["grade"] == "11th"].groupby(["school_name"]) ["math_score"].mean()
grade12 = merge.loc[merge["grade"] == "12th"].groupby(["school_name"]) ["math_score"].mean()
grade_summary = pd.DataFrame({"9th": grade9, "10th": grade10, "11th": grade11, "12th": grade12})
grade_summary
```

```python
read_grade9 = merge.loc[merge["grade"] == "9th"].groupby(["school_name"]) ["reading_score"].mean()
read_grade10 = merge.loc[merge["grade"] == "10th"].groupby(["school_name"]) ["reading_score"].mean()
read_grade11 = merge.loc[merge["grade"] == "11th"].groupby(["school_name"]) ["reading_score"].mean()
read_grade12 = merge.loc[merge["grade"] == "12th"].groupby(["school_name"]) ["reading_score"].mean()
read_summary = pd.DataFrame({"9th": read_grade9, "10th": read_grade10, "11th": read_grade11, "12th": read_grade12})
read_summary
```
